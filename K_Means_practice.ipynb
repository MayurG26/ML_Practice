```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

```python
iris_dataset = pd.read_csv("iris-dataset.csv")
```

```python
iris_dataset
```

```python
x = iris_dataset['sepal_length']
y = iris_dataset['sepal_width']
plt.scatter(x,y)
```

```python
k_means = KMeans(2)
data = iris_dataset.copy()
k_means.fit(data)
```

```python
cluster = iris_dataset.copy()
cluster["clusters"] = k_means.fit_predict(data)
plt.scatter(x,y,c=cluster["clusters"])
```

```python
from sklearn import preprocessing
x_scaled = preprocessing.scale(data)
```

```python
x_scaled
```

```python
k_means = KMeans(2)
k_means.fit(x_scaled)
```

```python
cluster = iris_dataset.copy()
cluster["clusters"] = k_means.fit_predict(x_scaled)
plt.scatter(x,y,c=cluster["clusters"])
```

```python
wcss = []
max_clusters = 10
for i in range(1,10):
    kmeans = KMeans(i)
    kmeans.fit(x_scaled)
    w = kmeans.inertia_
    wcss.append(w)
wcss
```

```python
num_of_cluster = range(1,10)
```

```python
num_of_cluster
```

```python
plt.plot(num_of_cluster,wcss)
plt.xlabel('Number of clusters')
plt.ylabel('Within-cluster Sum of Squares')
```
